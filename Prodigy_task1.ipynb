```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('world-population-by-country-2020.csv')
df1
```

```python
df1.info()
```

```python
df1.isna().sum()
```

```python
df1['Population 2020'] = df1['Population 2020'].str.replace(',', '')
df1['Population 2020'] = df1['Population 2020'].astype(float)
```

```python
df1.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df1['Population 2020'], bins=20)
plt.title("Histogram of Population (2020)")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
df_top10 = df1.sort_values('Population 2020', ascending=False).head(10)
plt.figure(figsize=(10,5))
plt.bar(df_top10['Country (or dependency)'], df_top10['Population 2020'])
plt.title("Top 10 Most Populated Countries (2020)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```
